# series

one dimensionalarray that is capable of storing various data types

```python
import pandas as pd
```

```python
x=[4,4,5,5,5,6]
var=pd.Series(x)
print(var)
```

```python
print(type(var))
```

```python
print(var[5])
```

```python
x=[4,5,6,7,8]
var=pd.Series(x,index=["a","b","c","d","e"])
print(var)
```

```python
x=[4,5,6,7,8]
var=pd.Series(x,index=["a","b","c","d","e"],dtype="float")
print(var)
```

```python
x=[4,5,6,7,8]
var=pd.Series(x,index=["a","b","c","d","e"],dtype="float",name="python")
print(var)
```

# data frame

```python
d=[1,2,3,4,5,6,7]
v=pd.DataFrame(d)
print(v)
```

```python
print(type(v))
```

```python
di={"a":[1,2,3,4,5],"b":[1,2,3,4,5]}
v1=pd.DataFrame(di)
print(type(v1))
print(v1)
```

```python
di={"a":[1,2,3,4,5],"b":[1,2,3,4,5],"c":[1,2,3,4,5]}
v1=pd.DataFrame(di,columns=["a","c"],index=["a","s","g","r","d"])
print(type(v1))
print(v1)
```
