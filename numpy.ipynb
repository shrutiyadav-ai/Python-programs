# Data types in array

```python
d=np.array([1,2,3,4])
print("data type:",d.dtype)
```

```python
d=np.array([1,2,3,4,45,67,77])
print("data type:",d.dtype)
```

```python
d=np.array([1.0,2.0,3.0,4.0])
print("data type:",d.dtype)
```

```python
d=np.array(["A","V","D","F","F"])
print("data type:",d.dtype)
```

```python
x=np.array([1,2,3,4])
print("data type:",x.dtype)
```

```python
x=np.array([1,2,3,4],dtype=np.int8)
print("data type:",x.dtype)
```

```python
x=np.array([1,2,3,4],dtype=np.int8)
print("data type:",x.dtype)
print(x)
```

```python
x1=np.array([1,2,3,4],dtype="f")
print("data type:",x1.dtype)
print(x1)
```

```python
x2=np.array([1,2,3,4])
new=np.float32(x2)
print("data type:",x2.dtype)
print("data type",new.dtype)
print(x2)
print(new)
```

```python
x2=np.array([1,2,3,4])
new=np.float32(x2)
new_1=np.int_(new)
print("data type:",x2.dtype)
print("data type",new.dtype)
print("data type",new_1.dtype)
print(x2)
print(new)
print(new_1)
```

```python
x3=np.array([1,2,3,4])
new_2=x3.astype(float)
print(x3)
print(new_2)
```

# RESHAPE

```python
s2=np.array([1,2,3,4,5,6])
x=s2.reshape(3,2)
print(x)
```

```python
s2=np.array([1,2,3,4,5,6])
print(s2)
print()

x=s2.reshape(3,2)
print(x)
```

```python
s2=np.array([1,2,3,4,5,6])
print(s2)
print()
print(s2.ndim)

x=s2.reshape(3,2)
print(x)
print(x.ndim)
```
